# Exploratory Data Analysis (EDA) and Data Cleaning

This notebook will guide you through:
- Importing and loading the dataset
- Cleaning and preprocessing the data
- Performing exploratory data analysis (EDA)
- Creating a focused, clean dataset for targeted analysis
- Visualizing insights with multiple graphs

Each step will be explained, including the purpose and interpretation of each graph.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('dataset/22070521075_CA1_EDA.csv')
df.head()
```

## Initial Data Inspection

Let's inspect the dataset to understand its structure, check for missing values, and identify columns for focused analysis.

```python
# Show basic info and missing values
df.info()
print('\nMissing values per column:')
print(df.isnull().sum())

# Show column names for selection
df.columns
```

## Data Cleaning and Focused Column Selection

Based on the inspection, we'll:
- Handle missing values (drop or fill as appropriate)
- Select a few key columns for focused analysis to improve accuracy and relevance

The cleaning process will be explained step by step.

```python
# Select focused columns for analysis
focused_columns = [
    'year', 'state_name', 'district_name',
    'rape_women_above_18', 'dowry_deaths',
    'kidnapping_and_abduction', 'cruelty_by_husband_or_his_relatives'
 ]

# Drop rows with missing values in focused columns
clean_df = df[focused_columns].dropna()

# Save clean dataset for further use
clean_df.to_csv('dataset/clean_dataset.csv', index=False)

# Show the cleaned, focused dataset
clean_df.head()
```

## Exploratory Data Analysis (EDA) and Visualization

We'll now perform EDA on the cleaned, focused dataset. Multiple graphs will be created to analyze distributions, relationships, and patterns. Each graph will be explained in detail.

```python
# Distribution of rape cases (women above 18)
plt.figure(figsize=(8,5))
sns.histplot(clean_df['rape_women_above_18'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Rape Cases (Women Above 18)')
plt.xlabel('Number of Cases')
plt.ylabel('Count')
plt.savefig('images/rape_women_above_18_hist.png')
plt.show()
print('This histogram shows the distribution of rape cases involving women above 18 across districts/years.')
```

```python
# Incidents by District
plt.figure(figsize=(10,6))
district_counts = clean_df['district_name'].value_counts().head(10)
sns.barplot(x=district_counts.values, y=district_counts.index, palette='viridis')
plt.title('Top 10 Districts by Incident Count')
plt.xlabel('Number of Incidents')
plt.ylabel('District')
plt.savefig('images/top10_districts_bar.png')
plt.show()
print('This bar plot shows the top 10 districts with the highest number of incidents, helping identify hotspots.')
```

```python
# Incident Type Distribution (using available columns)
incident_types = ['rape_women_above_18', 'dowry_deaths']
type_counts = clean_df[incident_types].sum()
plt.figure(figsize=(7,7))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Incident Type Distribution')
plt.savefig('images/incident_type_pie.png')
plt.show()
print('This pie chart shows the proportion of selected incident types (rape of women above 18 and dowry deaths), highlighting the most common types among those available.')
```

```python
# Incident counts by type and state (or district)
melted = clean_df.melt(id_vars=['state_name', 'district_name'], 
                      value_vars=['rape_women_above_18', 'dowry_deaths'],
                      var_name='Incident_Type', value_name='Count')
# Group by incident type and state (or district)
incident_counts = melted.groupby(['Incident_Type', 'state_name'])['Count'].sum().unstack().fillna(0)
ax = incident_counts.T.plot(kind='bar', stacked=False, colormap='tab20', figsize=(12,6))
ax.set_title('Incident Counts by Type and State')
ax.set_xlabel('State')
ax.set_ylabel('Number of Incidents')
plt.legend(title='Incident Type')
plt.tight_layout()
plt.savefig('images/incident_counts_by_state_bar.png')
plt.show()
print('This grouped bar plot compares the number of incidents for each type (rape of women above 18 and dowry deaths) across states.')
```

## Summary and Next Steps

- We loaded, cleaned, and focused the dataset for more accurate analysis.
- Multiple visualizations revealed key patterns in age, location, incident type, and outcomes.
- These insights can guide further analysis, predictive modeling, or targeted interventions.

You can now use the `clean_df` DataFrame for deeper analysis or modeling.

```python
# 1. Trend of incidents over years (line plot)
yearly_trend = clean_df.groupby('year')[['rape_women_above_18', 'dowry_deaths', 'kidnapping_and_abduction', 'cruelty_by_husband_or_his_relatives']].sum()
ax = yearly_trend.plot(marker='o', figsize=(10,6))
ax.set_title('Yearly Trend of Major Incident Types')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Incidents')
ax.grid(True)
plt.tight_layout()
plt.savefig('images/yearly_trend_line.png')
plt.show()
print("This line plot shows how the number of major incident types has changed over the years. It helps identify increasing or decreasing trends for each type.")

# 2. State-wise comparison of incidents (boxplot)
plt.figure(figsize=(14,6))
sns.boxplot(data=clean_df, x='state_name', y='rape_women_above_18')
plt.xticks(rotation=90)
plt.title('Distribution of Rape Incidents (Women Above 18) by State')
plt.xlabel('State')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.savefig('images/statewise_rape_boxplot.png')
plt.show()
print("This boxplot visualizes the spread and outliers of rape incidents across states, highlighting states with higher median or extreme values.")

# 3. Correlation heatmap between incident types
plt.figure(figsize=(8,6))
corr = clean_df[['rape_women_above_18', 'dowry_deaths', 'kidnapping_and_abduction', 'cruelty_by_husband_or_his_relatives']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Major Incident Types')
plt.tight_layout()
plt.savefig('images/incident_correlation_heatmap.png')
plt.show()
print("This heatmap shows the correlation between different incident types. Strong correlations may indicate related social issues or reporting patterns.")

# 4. Top 10 districts by cruelty incidents (horizontal bar plot)
plt.figure(figsize=(10,6))
top_cruelty = clean_df.groupby('district_name')['cruelty_by_husband_or_his_relatives'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_cruelty.values, y=top_cruelty.index, palette='mako')
plt.title('Top 10 Districts by Cruelty by Husband or Relatives')
plt.xlabel('Number of Incidents')
plt.ylabel('District')
plt.tight_layout()
plt.savefig('images/top10_cruelty_districts_bar.png')
plt.show()
print("This bar plot highlights the districts with the highest number of cruelty incidents, helping focus interventions on specific locations.")

# 5. Pairplot for incident types (scatter matrix)
sns.pairplot(clean_df[['rape_women_above_18', 'dowry_deaths', 'kidnapping_and_abduction', 'cruelty_by_husband_or_his_relatives']])
plt.suptitle('Pairwise Relationships Between Incident Types', y=1.02)
plt.savefig('images/incident_types_pairplot.png')
plt.show()
print("This pairplot shows the pairwise relationships between major incident types, revealing possible linear or non-linear associations.")
```
